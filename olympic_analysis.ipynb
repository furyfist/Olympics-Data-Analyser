The dataset is structured in an athlete-wise format, meaning each row corresponds to a single athlete. In the case of team events, such as hockey, every player on the team is listed individually and credited with the same medal. As a result, if a team of 11 players wins a gold medal, it appears as 11 gold medals in the dataset instead of one. This leads to overcounting when analyzing medal tallies for countries or events

To tackle this issue, we can identify and remove duplicate rows where all relevant column values—such as 'Team', 'Year', 'Event', and 'Medal'—are the same. This ensures that team events are counted only once per medal, rather than once per athlete. By dropping these duplicates, we prevent the overcounting of medals and obtain a more accurate representation of a country's total medal count.

Created a New Data Set Here Called medal_tally.

```python
medal_tally = df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])
```

```python
medal_tally = medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```
